###### Naive Bayes models are a group of extremely fast and simple classification algorithms that are often suitable for very high-dimensional datasets. Because they are so fast and have so few tunable parameters, they end up being very useful as a quick-anddirty baseline for a classification problem.

###### We would like to evaluate the model on data it has not seen before, and so we will split the data into a training set and a testing set.

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

#from sklearn.model_selection import load_iris

iris = load_iris()
X,y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB #Choosing model class
model = GaussianNB() #initiate model
model.fit(X_train, y_train) #fit model to data
y_model = model.predict(X_test) #predict on new data
print(y_model)
```

###### Finally, we can use the accuracy_score utility to see the fraction of predicted labels that match their true value

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_model)
accuracy_score
#print("Accuracy score: ", (accuracy_score * 100)+"%")
```
